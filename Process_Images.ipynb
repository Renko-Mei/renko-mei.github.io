```python
import os, sys
import numpy as np
import pandas as pd
from scipy.misc import imread, imshow, imresize
```

```python
# Read image data first (Oh this isn't really good for memory...)
preproc = lambda x: imresize(imread(x), [192, 256, 3])
folder = os.listdir('Images/')
folder.sort()
image_list = [preproc('Images/' + img) for img in folder]
image_data = pd.DataFrame({'DateTime':folder, 'Image':image_list})
image_data['Year'] = image_data['DateTime'].apply(lambda x: int(x[7:11]))
image_data['Month'] = image_data['DateTime'].apply(lambda x: int(x[11:13]))
image_data['Day'] = image_data['DateTime'].apply(lambda x: int(x[13:15]))
image_data['Hour'] = image_data['DateTime'].apply(lambda x: int(x[15:17]))
image_data = image_data.drop('DateTime', axis = 1)

# Remove super dark images
image_data['Brightness'] = image_data['Image'].apply(lambda x: np.mean(x))
image_data = image_data[image_data['Brightness'] > 50]
image_data.head(5)
```

```python
weather_data = pd.read_csv('Weather.csv')
data = weather_data.merge(image_data, how='left')
data = data.dropna()
data
```
